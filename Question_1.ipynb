```python
# Converting Year into no of Year
df['curr_Year']=2020
df['Years']=df.curr_Year - df.Year
df.drop(['Year','curr_Year'],axis=1,inplace=True)
df.head()
```

```python
# Making Dummies
df=pd.get_dummies(df,drop_first=True)
```

```python
df.head()
```

```python
# Plotting Corelation Between Columns Using Heatmap
plt.figure(figsize=(10,10))
sns.heatmap(df.corr(),annot=True,cmap='winter')
```

```python
# taking Dependent And Independent Column into Diff Variables
X= df.drop('Selling_Price',axis=1)
y=df['Selling_Price']
```

```python
print(X.shape)
print(y.shape)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=40)
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
X_train=scaler.fit_transform(X_train)
X_test=scaler.fit_transform(X_test)
```

```python
# Linear Regression and Logistic Regression
from sklearn.linear_model import LinearRegression, LogisticRegression
from sklearn.metrics import accuracy_score, classification_report
model = LinearRegression()
model.fit(X_train,y_train)
model.score(X_test,y_test)
```

```python
# Random Forest regression
from sklearn.ensemble import RandomForestRegressor
model = RandomForestRegressor()
model.fit(X_train,y_train)
model.score(X_test,y_test)
```
